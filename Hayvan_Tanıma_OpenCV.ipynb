```python
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
# Gerekli kütüphaneleri çağırdık.
```

```python
import numpy as np
# "Numpy" kütüphanesini çağırdık
```

```python
model = VGG16(weights = "imagenet" ) # "VGG16" bir sinir ağı mimairisdir. "model" adlı değişkenini oluşturduk ve ağırlığı "imagenet" olarak tanımladık.
```

```python
img_path = "cat.png" # "img_path" adlı değişkene tanımasını istediğimiz ve önceden yüklediğimiz hayvan resmini tanımlıyoruz.
img = image.load_img(img_path, target_size=(224,224)) # Gerekli boyuta dönüştürüyoruz.
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x , axis=0)
x = preprocess_input(x)
# Gerekli işlemleri yaptık.
```

```python
preds = model.predict(x) # Modeli denemek için resim atacağımız değişkeni tanımladık.
```

```python
preds 
```

```python
print(decode_predictions(preds, top=5)) # Hangi hayvana daha yakın olduğunu göstermesi için print komutunun içine "decode_predictions(preds, top=5)" yazdık.
# "top=5" ile en üstteki 5 değeri aldık.
```
